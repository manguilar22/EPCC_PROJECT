```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_json("../DATA/ImportantV2.json",orient="records")
df.dtypes
```

```python
df.head(3)
```

```python
sns.set_context("paper",font_scale=3); sns.set_style("darkgrid"); plt.figure(figsize=(12,12))
sns.kdeplot(df["Lat"],df["Lon"],cbar=True,legend=True,n_levels=1000)
```

```python
sns.set_context("poster")
plt.figure(figsize=(12,12))
sns.lmplot(data=df,x="Lat",y="Lon",order=2,
           line_kws={"lw":3,"color":"red"},size=20,legend=True)
```
