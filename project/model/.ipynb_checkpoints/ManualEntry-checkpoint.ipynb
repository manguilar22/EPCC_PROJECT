```python
import pandas as pd
import numpy as np
import pickle as p
import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict
from classes.OOP.Location import Location
%matplotlib inline
```

```python
data=[
    Location("Hospital Providence of El Paso",31,46,14.0,106,30,01.6),
    Location("Palmas Medical Center",31,46,12.3,106,29,56.6),
    Location("Kindred Hospital El Paso",31,46,43.4,106,28,38.1),
    Location("El Paso Speciality Hospital",31,46,43.4,106,28,39.0),
    Location("El Paso LTAC Hospital",31,47,01.6,106,28,27.3),
    Location("Las Palmas Rehabilitation Center",31,47,20.5,106,30,31.9),
    Location("Complete Care Community Hospital",31,48,03.6,106,30,36.1),
    Location("William Beaumont Army Medical Center",31,49,17.5,106,27,46.8),
    Location("University Medical Center Foundation",31,47,18.7,106,26,06.4),
    Location("Hospital of Providence East Campus",31,47,25.8,106,15,52.2),
    Location("Del Sol Medical Center",31,45,24.7,106,21,01.8)
]
```

```python
h=OrderedDict({
    "Name": list(map(lambda i: i.name,data)),
    "DMS": list(map(lambda i: i.DMS(),data)),
    "DD": list(map(lambda i: i.DD(),data)),
    "LatitudeDMS": list(map(lambda i: i.latitudeDMS(),data)),
    "LongitudeDMS": list(map(lambda i: i.longitudeDMS(),data)),
    "LatitudeDD": list(map(lambda i: i.latitudeDecimalDegree(),data)),
    "LongitudeDD": list(map(lambda i: i.longitudeDecimalDegree(),data)),
    "UUID": list(map(lambda i: i.UUID(),data))
})
df = pd.DataFrame(h)
```

```python
df.to_csv("../data/csv/ElPaso.csv",index=False)
```

```python
del df["DMS"]
del df["LatitudeDMS"]
del df["LongitudeDMS"]
#df.to_csv("../data/wolfram/WolframElPaso.csv",index=False)
```

```python
locations = dict(zip(df["DD"],df["Name"]))
locations
```

```python
A={ "start_name": list(map(lambda i: i.name,data)),
    "start_lat": list(map(lambda i: i.latitudeDecimalDegree(),data)),
    "start_lon": list(map(lambda i: i.longitudeDecimalDegree(),data))
  }
```

```python
pf=pd.DataFrame(A)
pf
```

```python
from classes.Algorithms.Algorithms import Algorithms
```

```python
A = list(zip(pf["start_lat"],pf["start_lon"]))
c = Algorithms().euclideanDistance(A)
```

```python
for a,b in c:
    print("From:\t{}\n\tTo:\t{}".format(a,b))
```

```python
start,end = zip(*c)
d = pd.DataFrame(np.matrix(end),columns=["end_lat","end_lon"])
pf["end_lat"] = d["end_lat"]
pf["end_lon"] = d["end_lon"]
```

```python
results = list(zip(pf["end_lat"],pf["end_lon"]))
pf["end_name"] = [locations[e] for e in results]
pf[["start_name","start_lat","start_lon","end_name","end_lat","end_lon"]]
```

```python
json = "../data/rest/SortedElPaso.json"
csv = "../data/csv/SortedElPaso.csv"
pf.to_json(json,orient="records") 
pf.to_csv(csv,encoding="UTF-8",index=False)
```
